Now the model will be fitted initially with the two most linearly related features as a simple linear regression. Then we will fit a multiple regression model taking in consideration more than one feature at a time.

```python
#data has been already shuffled by who made it available, we will also scale the dataset since 
#we are going to use it all and there's quite a difference in oprders of magnitude between features
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
scaler = preprocessing.MinMaxScaler()
```

```python
#data will be always shuffled at every iteration in order to get proper results from the test set
for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(data['ambient temp'], data['energy output per hour'],
                                                        test_size=0.25, shuffle=True)
    X_train = scaler.fit_transform(X_train.values.reshape(-1, 1))
    X_test = scaler.fit_transform(X_test.values.reshape(-1, 1))
    model_1 = LinearRegression().fit(X_train, y_train)
    print(model_1.score(X_test, y_test))
```

```python
for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(data['exhaust volume'], data['energy output per hour'],
                                                        test_size=0.25, shuffle=True)
    X_train = scaler.fit_transform(X_train.values.reshape(-1, 1))
    X_test = scaler.fit_transform(X_test.values.reshape(-1, 1))
    model_2 = LinearRegression().fit(X_train, y_train)
    print(model_2.score(X_test, y_test))
```

Fitting on ambient temperature gives the best R squared coefficient. Anyway, adding more parameters to the equation don't
really have worsening effects on R squared. Infact every parameter with worsening effects is multiplied by 0 in the equation. Adding more parameters can make the model capture trends better but, due to random chance, it can also induce the model to capture unrelated events which mistakenly improve R squared. For this reason, in the following will be used R squared adjusted to the number of parameters.

```python
def adj_R(score, observations, features):
    coeff = (observations-1)/(observations-features-1)
    return 1-((1-score)*coeff)
```

```python
for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(data[['ambient temp', 'exhaust volume']],
                                                        data['energy output per hour'],
                                                        test_size=0.25, shuffle=True)
    model_3 = LinearRegression().fit(X_train, y_train)
    model_score = model_3.score(X_test, y_test)
    print('regular:', model_score, 'adjusted:', adj_R(model_score, X_train.shape[0] , X_train.shape[1]))
```

```python
for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(data[['ambient temp', 'exhaust volume','ambient pressure']],
                                                        data['energy output per hour'],
                                                        test_size=0.25, shuffle=True)
    model_4 = LinearRegression().fit(X_train, y_train)
    model_score = model_4.score(X_test, y_test)
    print('regular:', model_score, 'adjusted:', adj_R(model_score, X_train.shape[0] , X_train.shape[1]))
```

```python
for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,0:-1], data['energy output per hour'],
                                                        test_size=0.25, shuffle=True)
    model_4 = LinearRegression().fit(X_train, y_train)
    model_score = model_4.score(X_test, y_test)
    print('regular:', model_score, 'adjusted:', adj_R(model_score, X_train.shape[0] , X_train.shape[1]))
```

Even adjusting R squared, regressing with all the data yelds better results in testing.
